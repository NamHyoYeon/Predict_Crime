```python
df_plcst_nm = df['PLCST_NM']
plsct_nm_list = df_plcst_nm.to_list()
plsct_nm_list = list(set(plsct_nm_list))
```

```python
plsct_nm_list
```

```python
plsct_nm_list_map = list()
for s in plsct_nm_list:
  plsct_nm_list_map.append(s)
```

```python
len(plsct_nm_list)
```

```python
for i in range(len(plsct_nm_list)):
  if plsct_nm_list[i] != 'ETC' and plsct_nm_list[i][-1] !='청' and plsct_nm_list[i][-1] != '과' and plsct_nm_list[i][-1] != '단':
    plsct_nm_list[i] = plsct_nm_list[i]+'경찰서'
  elif plsct_nm_list[i][-1] == '청' and plsct_nm_list[i] != '경찰청' and plsct_nm_list[i] != '산청':
    plsct_nm_list[i] =  plsct_nm_list[i][:-1]+'경찰청'
  elif plsct_nm_list[i] == '산청':
    plsct_nm_list[i] = '산청경찰서'
```

```python
for i in range(len(plsct_nm_list)):
  if plsct_nm_list[i] == '울산동부경찰서':
    plsct_nm_list[i] = '동부경찰서'
  elif plsct_nm_list[i] == '서울강남경찰서':
    plsct_nm_list[i] = '강남경찰서'
  elif plsct_nm_list[i] == '경상남도고성경찰서':
    plsct_nm_list[i] = '고성경찰서'
  elif plsct_nm_list[i] == '서울은평경찰서':
    plsct_nm_list[i] = '은평경찰서'
  elif plsct_nm_list[i] == '서울강남경찰서':
    plsct_nm_list[i] = '강남경찰서'
  elif plsct_nm_list[i] == '부산기장경찰서':
    plsct_nm_list[i] = '기장경찰서'
  elif plsct_nm_list[i] == '서울영등포경찰서':
    plsct_nm_list[i] = '영등포경찰서'
  elif plsct_nm_list[i] == '서울금천경찰서':
    plsct_nm_list[i] = '금천경찰서'
  elif plsct_nm_list[i] == '서울서대문경찰서':
    plsct_nm_list[i] = '서대문경찰서'
  elif plsct_nm_list[i] == '서울영등포경찰서':
    plsct_nm_list[i] = '영등포경찰서'
  elif plsct_nm_list[i] == '서울금천경찰서':
    plsct_nm_list[i] = '금천경찰서'
  elif plsct_nm_list[i] == '서울서대문경찰서':
    plsct_nm_list[i] = '서대문경찰서'
  elif plsct_nm_list[i] == '서울영등포경찰서':
    plsct_nm_list[i] = '영등포경찰서'
  elif plsct_nm_list[i] == '서울금천경찰서':
    plsct_nm_list[i] = '금천경찰서'
  elif plsct_nm_list[i] == '서울서대문경찰서':
    plsct_nm_list[i] = '서대문경찰서'
  elif plsct_nm_list[i] == '충남경찰청':
    plsct_nm_list[i] = '충청남도경찰청'
  elif plsct_nm_list[i] == '전남경찰청':
    plsct_nm_list[i] = '전라남도경찰청'
  elif plsct_nm_list[i] == '서울서대문경찰서':
    plsct_nm_list[i] = '서대문경찰서'
  elif plsct_nm_list[i] == '인천연수경찰서':
    plsct_nm_list[i] = '연수경찰서'
  elif plsct_nm_list[i] == '서울성동경찰서':
    plsct_nm_list[i] = '성동경찰서'
  elif plsct_nm_list[i] == '충북경찰청':
    plsct_nm_list[i] = '충청북도경찰청'
  elif plsct_nm_list[i] == '강원도고성경찰서':
    plsct_nm_list[i] = '고성경찰서'
```
